```python
Recommender System:- is a system that seeks to predict or filter preferences according to the user's choices.
                     Recommender systems are utilized in a variety of areas including movies, music, news,
                     books, research articles, search queries, social tags, and products in general.
                     Recommender systems produce a list of recommendations in any of the two ways:-

1. Collaborative filtering: Collaborative filtering approaches build a model from the user's past behavior
                            (i.e. items purchased or searched by the user) as well as similar decisions made
                            by other users.This model is then used to predict items (or ratings for items) that
                            users may have an interest in.

2.Content-based filtering: Content-based filtering approaches uses a series of discrete characteristics of an item in
                           order to recommend additional items with similar properties.Content-based filtering methods
                           are totally based on a description of the item and a profile of the user's preferences.

It recommends items based on the user's past preferences. Let's develop a basic recommendation system using Python and Pandas.
Let's develop a basic recommendation system by suggesting items that are most similar to a particular item, in this case, movies.
It just tells what movies/items are most similar to the user's movie choice.


```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score=cosine_similarity(X)
similarity_score
```

```python
similarity_score.shape
```

```python
favourite_movie_name=input('Enter your favourite movie name:')
```

```python
list_of_all_titles=df['Movie_Title'].tolist()
```

```python
import difflib
movie_recommendation=difflib.get_close_matches(favourite_movie_name,list_of_all_titles)
print(movie_recommendation)
```

```python
close_match=movie_recommendation[0]
print(close_match)
```

```python
index_of_close_match_movies=df[df.Movie_Title==close_match]['Movie_ID'].values[0]
print(index_of_close_match_movies)
```

```python
#getting a score of similar movies
recommendation_score=list(enumerate(similarity_score[index_of_close_match_movies]))
print(recommendation_score)
```

```python
len(recommendation_score)
```

```python


#model evaluation

```

```python
sorted_similar_movies=sorted(recommendation_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
print('Top 30 movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=df[df.index==index]['Movie_Title'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```

```python


#prediction

```

```python
movie_name=input('Enter your favourite movie name:')
list_of_all_titles=df['Movie_Title'].tolist()
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)

close_Match=find_close_match[0]
index_of_movie=df[df.Movie_Title==close_Match]['Movie_ID'].values[0]
recommendation_score=list(enumerate(similarity_score[index_of_movie]))
sorted_similar_movies=sorted(recommendation_score,key=lambda x:x[1],reverse=True)

print('Top 10 movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=df[df.index==index]['Movie_Title'].values[0]
  if(i<11):
    print(i,'.',title_from_index)
    i+=1
```

# Project Overview
This notebook implements a movie recommendation system using a dataset that includes various attributes of movies. The goal is to recommend movies to users based on the similarity of movies they like to other movies in the dataset.


# Dataset
 The dataset used contains the following columns:

 \

Movie_ID: A unique identifier for each movie.

Movie_Title: The title of the movie.

Movie_Genre: The genre of the movie.

Movie_Language: The language in which the movie is produced.

Movie_Budget: The budget of the movie.

Movie_Popularity: A measure of the movie's popularity.

Movie_Release_Date: The release date of the movie.

Movie_Revenue: The revenue generated by the movie.

Movie_Runtime: The runtime of the movie.

Movie_Vote: The average vote the movie received.

Movie_Vote_Count: The number of votes the movie received.

Movie_Homepage: The official homepage of the movie.

Movie_Keywords: Keywords associated with the movie.

Movie_Overview: A brief overview of the movie.

Movie_Production_House: The production house that produced the movie.

Movie_Production_Country: The country where the movie was produced.

Movie_Spoken_Language: The language spoken in the movie.

Movie_Tagline: The tagline of the movie.

Movie_Cast: The main cast of the movie.

Movie_Crew: The crew members of the movie.

Movie_Director: The director of the movie.

\
# Methodology
Data Loading: The dataset is loaded and preprocessed to ensure it is ready for analysis.
Similarity Score Calculation: A similarity score is calculated using cosine similarity on the relevant features of the movies. This score helps in identifying movies that are similar to each other.
User Input: The user is prompted to enter the name of a movie they like.
Recommendation Generation: Based on the user's input, the system finds the closest matching movie in the dataset and recommends a list of similar movies. The recommendations are sorted by their similarity score in descending order.

\
# Code Implementation
Data Loading: The code loads the dataset using pandas and performs any necessary cleaning and preprocessing.
Similarity Calculation: The cosine similarity between movies is calculated using relevant features from the dataset.
User Interaction: The system takes a movie name as input from the user, finds the closest match, and generates a list of recommended movies.

\
# Results
The notebook outputs the top 10 movie recommendations based on the user's input. For example, if the user enters "Toy Story," the system may recommend movies like "Toy Story 2," "Mulan," and "End of Days."

\
# Conclusion
This project demonstrates a basic implementation of a movie recommendation system using similarity scores. The recommendations are based on the assumption that movies with similar attributes are likely to be enjoyed by the same audience.
